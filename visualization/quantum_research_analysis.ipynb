## Quantum Research Trends Case Study

Exploratory Data Analysis on arXiv 'quant-ph' papers.

To Understand publication trends and keyword patterns in quantum computing research.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df_keywords = pd.read_csv('C:/Users/vinay/top_keywords.csv')
df_years = pd.read_csv('C:/Users/vinay/publications_by_year.csv', index_col=0)


df_years.columns = ['count']
df_years.head()
```

```python
plt.figure(figsize=(10, 5))
df_years.plot(marker='o', legend=False)
plt.title("Quantum Physics Publications per Year")
plt.xlabel("Year")
plt.ylabel("Number of Papers")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top20 = df_keywords.head(20)
plt.barh(top20['word'][::-1], top20['frequency'][::-1])
plt.title("Top 20 Keywords in Quantum Research in Titles and abstracts")
plt.xlabel("Frequency")
plt.tight_layout()
plt.show()
```
